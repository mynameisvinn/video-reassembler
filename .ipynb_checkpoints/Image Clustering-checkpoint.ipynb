```python
from skimage.io import imread
from skimage import transform as tf
from skimage.feature import (match_descriptors, corner_harris, corner_peaks, ORB, plot_matches)
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
import glob
from skimage.color import rgb2hsv
import time
import shutil
%matplotlib inline
```

```python
class node:
        
    def __init__(self, path):
        self.path = path
        self.image = imread(path)
            
        
    def calculate_hue_histogram(self, bins):
        '''
        function
        --------
        generate color histogram since color seems to be a discriminating feature in scenes.
        
        parameters:
        -----------
        @bins: int, refers to number of bins
        '''
        self.hue_hist = np.histogram(rgb2hsv(self.image)[0], bins, density = True)[0] # 0 channel is hue, [0] is value counts (1) is bin value
    
    
    
    
    def calculate_lbp(self, num_keypoints):
        '''
        function
        --------
        extract lbp features from image.
        
        parameters
        ----------
        @num_keypoints: int, refers to number of keypoints to extract
        '''
        descriptor_extractor = ORB(n_keypoints = num_keypoints)        
        image_gray = rgb2gray(self.image) # lbp is intensity not color
        descriptor_extractor.detect_and_extract(image_gray)
        self.lbp = pd.DataFrame(descriptor_extractor.descriptors)

        
        
    def get_lbp(self):
        '''
        function:
        ---------
        called by fit(), predict() and transform(). it's necessary for downstream modification
        of lbp features such as pca reduction and kmeans mapping
        '''
        return self.lbp
    
    
    
    def get_image(self):
        return self.image
        
        
        
    def add_word(self, words):
        '''
        function
        --------
        after lbp is mapped to a word by transform(), it needs to be saved to the 
        corresponding node.
        
        parameters:
        -----------
        @words: visual word df generated by transform()'s kmeans prediction
        '''
        
        self.words = words
        
        
        
        
    def words_to_histogram(self, num_words):
        '''
        function
        --------
        converts words into histograms for comparison
        
        returns
        -------
        @self.lbp_hist_sorted: histogram of lbp words
        '''
        
        bin_values = [x for x in range(num_words)] # number of bins should equal number of visual  words
        lbp_hist_unsorted = pd.Series(self.words).value_counts().reindex(bin_values, fill_value=0)
        self.lbp_hist_sorted = lbp_hist_unsorted.sort_index().values
        
        
        
    def merge_hue_words(self):
        '''
        function
        --------
        merge_hue_words() merges lbp histogram and hue histogram
        
        returns
        -------
        @self.merged_hist: merged histogram 
        '''
        self.merged_hist = np.hstack([self.lbp_hist_sorted, self.hue_hist])
```

```python
class image_cluster:
    
    def __init__(self):
        pass
                 
                 
    def fit(self, directory_path, num_words, num_keypoints, pca_dim):
        '''
        function
        --------
        fit() instantiates reducer, kmeans, etc.
        
        parameters
        ----------
        @directory_path: string. path to training folder.
        @num_words: int. number of visual words
        @num_keypoints: int. number of lbp keypoints.
        @pca_dim: number of pca dimensions to retain.
        '''
        
        # track hyperparameters
        list_image_paths = glob.glob(directory_path)
        self.dict_nodes = {} # track training images
        self.num_words = num_words # used by transform()
        self.num_keypoints = num_keypoints # used by predict()
        
        # create dictionary of nodes
        for idx, path in enumerate(list_image_paths):
            self.dict_nodes[idx] = node(path)
            
        # have each node generate lbp features; then store lbp features in a master list
        lbp_list = []
        
        for idx in range(len(self.dict_nodes)):
            self.dict_nodes[idx].calculate_lbp(num_keypoints)
            lbp_list.append(self.dict_nodes[idx].get_lbp())
        
        # convert list of lbp features to dataframe
        lbp_df = pd.DataFrame(np.vstack(lbp_list))
        
        # fit & transform reducer
        self.reducer = PCA(n_components=pca_dim)
        lbp_reduced_df = self.reducer.fit_transform(lbp_df)
        
        # fit & train kmeans
        self.k_means = KMeans(num_words, random_state = 1) # seed for reproducibility
        self.k_means.fit(lbp_reduced_df)
    

    
    def transform(self, num_bins):
        
        '''
        function
        --------
        transform() converts each image to a word histogram, using pca + kmeans
        
        parameters:
        -----------
        @num_bins: int, refers to number of bins for color histogram
        '''
        
        self.num_bins = num_bins # num_bins will be used again by predict()
        
        
        # tell each node to convert lbp features into words
        for idx in range(len(self.dict_nodes)):
            
            # grab lbp features from each node
            lbp_df = self.dict_nodes[idx].get_lbp()
            
            # reduce each lbp feature
            lbp_reduced_df = self.reducer.transform(lbp_df)
            
            # map reduced lbp to word
            word_df = self.k_means.predict(lbp_reduced_df)
            
            # save it to the node
            self.dict_nodes[idx].add_word(word_df)
            
            
        # then, tell each node to convert words to a single histogram
        for idx in range(len(self.dict_nodes)):
            self.dict_nodes[idx].words_to_histogram(self.num_words)
            
        # tell each node to generate a color histogram and then merge it with lbp hist
        for idx in range(len(self.dict_nodes)):
            self.dict_nodes[idx].calculate_hue_histogram(self.num_bins)
            self.dict_nodes[idx].merge_hue_words()
            
        
        
    def predict(self, path):
        '''
        function:
        ---------
        predict() finds nearest neighbors when given an image. distance is calculated using the chi squared method.
        
        
        parameters:
        -----------
        @path: image path
        '''
        
        # create test node 
        test_node = node(path)
        
        # tell test node to generate lbp features
        test_node.calculate_lbp(self.num_keypoints)
        
        # grab lbp features from each node
        test_lbp_df = test_node.get_lbp()
        
        # reduce lbp features
        test_lbp_reduced_df = self.reducer.transform(test_lbp_df)
        
        # map lbp features to visual words
        test_words_df = self.k_means.predict(test_lbp_reduced_df)
        
        # then store visual words for each node
        test_node.add_word(test_words_df)
        
        # convert visual words to a histogram
        test_node.words_to_histogram(self.num_words)
        
        # first generate color histogram, then merge it with lbp features
        test_node.calculate_hue_histogram(self.num_bins) # num_bins defined during transform() function
        test_node.merge_hue_words()
        
        # find nearest neighbor by generating a list of tuples (idx, distances)
        self.list_distances = []
        
        for idx in range(len(self.dict_nodes)):
                        
            # generate distances and store it in a list
            self.list_distances.append((idx, self._chi2_distance(test_node.merged_hist, self.dict_nodes[idx].merged_hist)))
            
        # sort by the 2nd element in the tuple, which correspond to distance, and return top three closest matches
        return sorted(self.list_distances, key=lambda tup: tup[1])
                
        
    def generate_histogram_features(self, path):
        '''
        function:
        ---------
        extract feature vector
        '''
        
        # create test node 
        test_node = node(path)
        
        # tell test node to generate lbp features
        test_node.calculate_lbp(self.num_keypoints)
        
        # grab lbp features from each node
        test_lbp_df = test_node.get_lbp()
        
        # reduce lbp features
        test_lbp_reduced_df = self.reducer.transform(test_lbp_df)
        
        # map lbp features to visual words
        test_words_df = self.k_means.predict(test_lbp_reduced_df)
        
        # then store visual words for each node
        test_node.add_word(test_words_df)
        
        # convert visual words to a histogram
        test_node.words_to_histogram(self.num_words)
        
        # first generate color histogram, then merge it with lbp features
        test_node.calculate_hue_histogram(self.num_bins) # num_bins defined during transform() function
        test_node.merge_hue_words()
        
        # return histogram for kmeans clustering
        return test_node.merged_hist
                
                
    def _chi2_distance(self, histA, histB, eps = 1e-10):
        '''
        function:
        ---------
        called by predict(). chi2_distance is the similarity function used to compare 
        histograms.
        
        parameters:
        -----------
        @histA: numpy array
        @histB: numpy array
        
        returns:
        --------
        @distance: float, corresponding to distance between histA and histB
        '''
        distance = 0.5 * np.sum([((a - b) ** 2) / (a + b + eps) for (a, b) in zip(histA, histB)])
        return distance
        

        
    def show_image(self, idx):
        '''
        function:
        ---------
        display image when given index.
        '''
        plt.imshow(self.dict_nodes[idx].get_image())
```
